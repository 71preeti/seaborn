RELATIONAL PLOT MEANS relplot() = by default create scatter plot<br>
 isse do plot he bna skte h <br>
1. scatter plot <br>
2. line plot

```python
sns.relplot(x="tip",y="total_bill",data=df, hue="sex",style="smoker")
plt.show()               # this plot show positive relation
```

```python
sns.relplot(x="tip",y="total_bill",data=df, hue="sex",style="smoker",col='day')
plt.show()  
```

```python
sns.relplot(x="tip",y="total_bill",data=df, hue="time",style="smoker",col='day')
plt.show()  
```

```python
sns.relplot(x="total_bill",y="size",data=df)  # it also show positive relation
plt.show()
```

```python
df.dtypes
```

```python
df[['total_bill','tip','size']]
```

```python
df[['total_bill','tip','size']].corr()  # it tells about correlation among data
```

```python
sns.relplot(x="tip",y="total_bill",data=df,hue="sex",kind="line")
plt.show                       # line plot 
```

CATEGORICAL PLOT AS catplot() <br>
used for one numerical and other categorical column <br>
by default it show strip plot 


```python
sns.catplot(x="smoker",y="total_bill",data=df)
plt.show()             # it show strip plot by default
```

```python
sns.catplot(x="smoker",y="tip",data=df,hue="sex")
plt.show()
```

```python
sns.catplot(x="smoker",y="total_bill",data=df,hue="sex",kind="violin")
plt.show()
```

```python
sns.catplot(x="smoker",y="total_bill",data=df,hue="sex",kind="swarm")
plt.show()
```

```python
sns.catplot(x="smoker",y="total_bill",data=df,hue="sex",kind="box")
plt.show()
```

```python
sns.catplot(x="smoker",y="total_bill",data=df,hue="sex",kind="boxen")
plt.show()
```

```python
sns.catplot(x="smoker",data=df,hue="sex",kind="count")  # isme ek axis he pass ho skti h 
plt.show()                        # means 1 column ko dene ki need hoti h bss
```

```python
sns.catplot(x="smoker",y="total_bill",data=df,hue="sex",kind="point")
plt.show()
```

```python
sns.displot(df["total_bill"],kde=True,color="r")        # it is same as  histogram, distrinution plot
plt.title("HISTOGRAM WITH KDE",color="g")
plt.show
```

```python
sns.histplot(df["total_bill"],color="g")
```

```python
sns.kdeplot(df["total_bill"])      # kde= kernel density estimate , isse hum graph distribution ko dekhte h ki ye normal distributed h ya nhi  
plt.show()           # it is like normal distribution . so Missing value NaN  filled hogi with "MEAN"
```

```python
sns.kdeplot(df["size"])             # it is not normal distribution . so missing NaN value filled by "MEDIAN".
plt.show()
```
